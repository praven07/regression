```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#housing data. x = size, y = price
data = np.array([
    [1.0, 300.0],
    [2.0, 500.0],
])
plt.scatter(data[: , 0], data[:, 1])
```

```python
#split the data for training
x_train = data[:, 0]
y_train = data[:, 1]
```

```python
def compute_cost(x: np.array, y: np.array, w, b):
    
    m = x.shape[0] #number of training examples
    
    y_hat = w * x + b
    
    cost_sum = np.sum(((y_hat - y) ** 2))
    total_cost = (1 / (2 * m)) * cost_sum
    
    return total_cost
```

```python
def predict(w, b, x):
    
    return w * x + b
```

```python
weight = 200
bias = 100

cost = compute_cost(x_train, y_train, weight, bias)
print(f"Cost Function: {cost}")
plt.scatter(x_train, y_train)
plt.plot(x_train, predict(weight, bias, x_train), color='red')
```
